```python
import pandas as pd
import numpy as np


csv = pd.read_csv('water_counties_cleaned.csv')
csv.head()
```

```python

northeast_div1 = ['MA', 'NH', 'VT', 'RI', 'CT', 'ME']
northeast_div2 = ['PA', 'NY', 'NJ']

midwest_div1 = ['IL', 'IN', "MI", 'OH', 'WI']
midwest_div2 = ['IA', 'KS', 'MN', 'MO', 'NE', 'ND', 'SD']

south_div1 = ['DE', 'FL', 'GA', 'MD', 'NC', 'MS', 'SC', 'VA', 'DC', 'WV']
south_div2 = ['AL', 'KY', 'MI', 'TN']
south_div3 = ['AR', 'LA', 'OK', 'TX']

west_div1 = ['AZ', 'CO', 'ID', 'MT', 'NV', 'NM', 'UT', 'WY']
west_div2 = ['AK', 'CA', "WA", 'OR']

islands = ['HI', 'PR', 'VI']


def getRegion(state):
    if state in northeast_div1 + northeast_div2:
        return 'Northeast'
    elif state in midwest_div1 + midwest_div2:
        return 'Midwest'
    elif state in south_div1 + south_div2 + south_div3:
        return 'South'
    elif state in west_div1 + west_div2:
        return 'West'
    elif state in islands:
        return 'Islands'
```

```python
full = pd.read_csv("usco2015v2.0.csv")
full.head()
```

```python
# We care about "
"""
STATE	
Tp-TotPop	* 1000
COUNTY	
PS-GWPop	* 1000000
DO-PSDel	* 1000000
IN-WGWFr	* 1000000
IR-WGWFr	* 1000000
LI-WGWFr	* 1000000
PT-WGWFr	* 1000000
"""

selected = ['STATE', 'COUNTY']
selectedX1e3 = ['TP-TotPop']
selectedX1e6 = ['PS-GWPop', 'DO-PSDel', 'IN-WGWFr', 'IR-WGWFr', 'LI-WGWFr', 'PT-WGWFr']


full[selectedX1e6] = full[selectedX1e6].astype('int32')

new_df = full[selected]
new_df[['state', 'name']] = new_df[selected]
new_df['totpop'] = full[selectedX1e3] * 1e3
new_df[['gwpop', 'domestic', 'industrial', 'irrigation', 'livestock', 'thermoelectric']] = full[selectedX1e6] * 1e6
```

```python
by_state = new_df.groupby(['state']).sum().to_dict(orient='index')

new_df['region'] = new_df['state'].apply(getRegion)
by_region = new_df.groupby(['region']).sum().to_dict(orient='index')
```

```python
regions = {}

def toDict(val):
    return {
        "name": val['name'],
        "value": val['totpop'],
        "info": val
    }

for state in full['STATE'].unique():
    
    state_data = new_df[new_df.state == state]
    state_dict = state_data.to_dict(orient='records')
    dict_state_data = list(map(toDict, state_dict))
    reg = getRegion(state)
    
    if reg in regions:
        regions[reg]['children'].append({
            "name": state,
            "info": by_state[state],
            "children": dict_state_data
        })
    else: 
        regions[reg] = {
            "name": reg,
            "info": by_region[reg],
            "children" : [{
                "name": state,
                "info": by_state[state],
                "children": dict_state_data
            }]
        }

    
region_list = []
for reg in regions.keys():
    region_list.append(regions[reg])
    
with open('./docs/output.json', 'w') as output:
    string = str(region_list)
    string = string.replace("'", '"')
    string = string.replace('O"B', "O'B")
    string = string.replace('"s ', "'s ")
    output.write(string)
```
